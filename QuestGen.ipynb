```
from nltk.corpus import wordnet as wn

def explore_word_senses(word):
    synsets = wn.synsets(word, 'n')
    if not synsets:
        print(f"No senses found for the word '{word}'.")
        return
    print(f"Senses for the word '{word}':")
    print("-" * 40)

    for syn in synsets:
        print(f"Synset: {syn.name()}")
        print(f"Definition: {syn.definition()}")
        # Hypernyms
        hypernyms = syn.hypernyms()
        if hypernyms:
            print("Hypernyms:")
            for hypernym in hypernyms:
                print(f"  - {hypernym.name()}: {hypernym.definition()}")
        else:
            print("No Hypernyms found.")

        # Hyponyms
        hyponyms = syn.hyponyms()
        if hyponyms:
            print("Hyponyms:")
            for hyponym in hyponyms:
                print(f"  - {hyponym.name()}: {hyponym.definition()}")
        else:
            print("No Hyponyms found.")
        # Examples
        examples = syn.examples()
        if examples:
            print("Examples:")
            for example in examples:
                print(f"  - {example}")
        else:
            print("No Examples found.")

        print("-" * 40)
```

```
original_word = "software"
explore_word_senses(original_word)
```

## Distractors Generation

```
from nltk.corpus import wordnet as wn

def get_distractors_wordnet2(syn, original_word):
    distractors = []

    # Get hypernyms for the given synset
    hypernym = syn.hypernyms()

    if not hypernym:
        return distractors

    hyponyms = hypernym[0].hyponyms()

    # Get hyponyms for each hypernym
    for hyponym in hyponyms:
        name = hyponym.lemmas()[0].name()
        name = name.replace("_"," ")
        if name not in distractors and name.capitalize() != original_word.capitalize():
          distractors.append(name.capitalize())

    return distractors
```

## Word Sense Disambiguation with Lesk Algorithm

In this cell, we implement the Lesk algorithm to perform Word Sense Disambiguation (WSD). The Lesk algorithm is a widely-used method for determining the most appropriate sense of a word in a given context based on the overlap of word sense definitions with the surrounding context. The objective is to resolve the ambiguity of a word and identify its intended meaning in a specific context.

```
from pywsd.lesk import simple_lesk
def get_wsd(sent,ambi):
  answer = simple_lesk(sent, ambi, pos='n')
  if answer is not None:
   return (answer,answer.definition(),ambi)
  return None
```

```
from pywsd.lesk import simple_lesk
from nltk.corpus import wordnet as wn

def get_wsd_details(sentence, ambiguous_word):
    answer = simple_lesk(sentence, ambiguous_word, pos='n')

    if answer is not None:
        sense_info = {
            'Synset': answer.name(),
            'Defn': answer.definition(),
            'Example': answer.examples()[0],
            'POS': answer.pos(),
            'Lemmas': [lemma.name() for lemma in answer.lemmas()],
        }

        return (sense_info, ambiguous_word)

    return (None, ambiguous_word)
```

```
# Example usage
sentence_example = "The bank by the river is a peaceful spot."
ambiguous_word_example = "bank"

result_example, word_example = get_wsd_details(sentence_example, ambiguous_word_example)
print(f"\nSense information for the word '{word_example}':\n")
print(result_example)
```

## Generate a question using context and answer with T5

```
from transformers import T5ForConditionalGeneration,T5Tokenizer

question_model = T5ForConditionalGeneration.from_pretrained('ramsrigouthamg/t5_squad_v1')
question_tokenizer = T5Tokenizer.from_pretrained('t5-base')

def get_question(sentence,answer):
  text = "context: {} answer: {} </s>".format(sentence,answer)
  print (text)
  max_len = 256
  encoding = question_tokenizer.encode_plus(text,max_length=max_len, pad_to_max_length=True, return_tensors="pt")

  input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]

  outs = question_model.generate(input_ids=input_ids,
                                  attention_mask=attention_mask,
                                  early_stopping=True,
                                  num_beams=5,
                                  num_return_sequences=1,
                                  no_repeat_ngram_size=2,
                                  max_length=200)


  dec = [question_tokenizer.decode(ids) for ids in outs]


  Question = dec[0].replace("question:","")
  Question= Question.strip()
  return Question
```

```
from sentence_transformers import SentenceTransformer, util
import textstat

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

def getMCQs(sent,questlist,q=5):
  keys = get_keywords_yakes(sent)
  tot_sim_score = 0
  tot_grade_score = 0
  count = 0
  for key in keys:
    if(count == q):
      break
    if(len(key.split()) == 1):
      ops = get_wsd(sent,key)
      if ops is not None:
        sense = ops[0]
        meaning = ops[1]
        answer = ops[2]
        if sense is not None:
          distractors = get_distractors_wordnet2(sense,answer)
          if(len(distractors) > 1):
            question = get_question(sent,answer)
            questlist.append(question)
            print("\n")
            print ("Question ->",question)
            print ("Answer ->",answer)
            print ("Distractors ->",distractors)
            #print ("Sense ->" ,sense)
            #print ("Meaning ->",meaning)
            #print("\n")
            count = count+1
        else:
          distractors = ["Word not found in Wordnet. So unable to extract distractors."]
      else:
        answer = key
        distractors = get_distractors_sense2vec(answer)
        if distractors != None:
          question = get_question(sent,answer)
          print("\n")
          print ("Question ->",question)
          print ("Answer ->",answer)
          print ("Distractors ->",distractors)
          #print ("Sense ->" ,sense)
          #print ("Meaning ->",meaning)
          #print("\n")
          count = count+1
    else:
      answer = key
      distractors = get_distractors_sense2vec(answer)
      if distractors != None:
        question = get_question(sent,answer)
        print("\n")
        print ("Question ->",question)
        print ("Answer ->",answer)
        print ("Distractors ->",distractors)
        #print ("Sense ->" ,sense)
        #print ("Meaning ->",meaning)
        #print("\n")
        count = count+1

    embedding_1= model.encode(question, convert_to_tensor=True)
    embedding_2 = model.encode(sent, convert_to_tensor=True)

    similarity_scores = util.pytorch_cos_sim(embedding_1, embedding_2)

    # Extract the similarity score as a float
    similarity_score = float(similarity_scores[0][0])
    tot_sim_score += similarity_score
    print("Similarity score ->",similarity_score)
    grade_level = textstat.flesch_kincaid_grade(question)
    tot_grade_score += grade_level
    print("Grade Level ->",grade_level)

    print("\n")
  print("\n\n Average Similarity Score :",float(tot_sim_score/q))
  print("\n\n Average Grade levels :",float(tot_grade_score/q))
```
